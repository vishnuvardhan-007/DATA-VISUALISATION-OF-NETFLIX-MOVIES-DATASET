```python
df['release_year'].value_counts()
```

```python
plt.figure(figsize = (16,9))
df['release_year'].value_counts().sort_index().plot(kind='bar', color='skyblue');
plt.title("Released Year Distribution");
plt.xlabel("Released Year");
plt.ylabel("Count");
plt.grid()
```

```python
sns.countplot(data=df, x='type', palette='viridis')
plt.title('Content Type Distribution')
plt.show()
```

```python
plt.pie(df['type'].value_counts() , labels = df['type'].value_counts().index,
       autopct = '%.2f%%', startangle = 0, shadow = True);
```

```python
sns.countplot(data=df, x='rating', hue='type', order=df['rating'].value_counts().index[:10])
plt.title('Ratings by Content Type')
plt.xticks(rotation=45)
plt.show()
```

```python
df['director'].value_counts()[0:12]
```

```python
sns.barplot(x=df['director'].value_counts()[1:11].index,y=df['director'].value_counts()[1:11].values)
plt.xticks(rotation=90)
plt.title(' Top 10 Directors')
plt.yticks(np.arange(1,22))
plt.ylabel('counter')
plt.show()
```

```python
df['country'].value_counts()[0:12]
```

```python
plt.pie(df['country'].value_counts()[0:11].values,labels=df['country'].value_counts()[0:11].index,explode=[0,0,0.4,0,0,0,0,0,0,0.1,0],autopct="%0.0f%%",shadow=True)
plt.title('Most of the countries where the TV shows  and movies were produced')
plt.show()
```

```python
plt.pie(df['cast'].value_counts()[0:11].values,labels=df['cast'].value_counts()[0:11].index,explode=[0.4,0,0.4,0,0,0,0,0,0,0.1,0],autopct="%0.0f%%",shadow=True)
plt.show()
```

```python
sns.barplot(x=df['cast'].value_counts()[1:11].index,y=df['cast'].value_counts()[1:11].values)
plt.xticks(rotation=90)
plt.title(' No of movies Done')
plt.yticks(np.arange(1,22))
plt.ylabel('number')
plt.show()
```

```python
sns.barplot(x=df['cast'].value_counts()[1:11].index,y=df['cast'].value_counts()[1:11].values)
plt.xticks(rotation=90)
plt.title(' No of movies Done')
plt.yticks(np.arange(1,22))
plt.ylabel('number')
plt.show()
```

```python
tv_shows = df[df['type'] == 'TV Show']
tv_shows['seasons'] = tv_shows['duration'].str.extract(r'(\d+)').astype(float)
plt.hist(tv_shows['seasons'].dropna(), bins=10, color='purple', edgecolor='black')
plt.title('Distribution of TV Shows by Seasons')
plt.xlabel('Number of Seasons')
plt.ylabel('Count')
plt.show()
```
